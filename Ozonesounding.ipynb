```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure numeric values and clean data
df['Temp'] = pd.to_numeric(df['Temp'], errors='coerce')
df['O3_mPa'] = pd.to_numeric(df['O3_mPa'], errors='coerce')
df['GeopAlt'] = pd.to_numeric(df['GeopAlt'], errors='coerce')
df = df.dropna(subset=['Temp', 'O3_mPa', 'GeopAlt'])

# Remove extreme O3 outliers if needed (e.g., > 15 mPa)
df = df[df['O3_mPa'] < 15]

# Sort by altitude
df = df.sort_values('GeopAlt')

# Create the plot
plt.figure(figsize=(10, 8))

# Plot both temperature and ozone on same x-axis
plt.plot(df['Temp'], df['GeopAlt'], color='red', linestyle='--', label='Temperature (°C)')
plt.plot(df['O3_mPa'], df['GeopAlt'], color='blue', linestyle='-', label='O₃ (mPa)')

# Axis labels and title
plt.xlabel('(°C or mPa)')
plt.ylabel('Geopotential Altitude (km)')
plt.title('Temperature and Ozone Vertical Profile')

# Make x-axis symmetrical
x_max = max(abs(df['Temp'].min()), abs(df['Temp'].max()), abs(df['O3_mPa'].min()), abs(df['O3_mPa'].max()))
plt.xlim(-x_max, x_max)

# Grid, legend, and layout
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
